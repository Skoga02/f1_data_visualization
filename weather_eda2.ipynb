```python
import duckdb
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

conn = duckdb.connect("f1_weather_only.duckdb")

all_weather_query = """
    SELECT 
        *
    FROM weather
    JOIN session USING (session_key)
    ORDER BY date
"""



all_weather_df = conn.execute(all_weather_query).df()


conn.close()

print(len(all_weather_df))
```

### Simple EDA

```python
all_weather_df.info()
```

## Vilket år hade högst average temperatur på alla banor?

```python
hottest_year = all_weather_df.groupby("year")["air_temperature"].mean().sort_values(ascending=False)

print(hottest_year)
print(f"Genomsnittligt varmaste året: {hottest_year.index[0]}")
```

### Top 5 Varmaste asfalt temperaturer

```python
hottest_tracks = all_weather_df.groupby("circuit_short_name")["track_temperature"].mean().sort_values(ascending=False).head(5)

print(hottest_tracks)
```

### Extra Matplotlib graf

```python
grouped_temperature_df = all_weather_df.groupby('circuit_short_name')[['air_temperature', 'track_temperature']].mean().reset_index()


grouped_temperature_df = grouped_temperature_df.sort_values(by='track_temperature', ascending=True)

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(10, 8))

ax.hlines(y=grouped_temperature_df['circuit_short_name'],
        xmin=grouped_temperature_df['air_temperature'],
        xmax=grouped_temperature_df['track_temperature'],
        color='gray',
        alpha=0.6,
        linewidth=2,
        zorder=1)

ax.scatter(grouped_temperature_df['air_temperature'],
        grouped_temperature_df['circuit_short_name'],
        color='lightblue',
        label='Air Temperature (°C)',
        s=100,
        zorder=2)

ax.scatter(grouped_temperature_df['track_temperature'],
        grouped_temperature_df['circuit_short_name'],
        color='orange',
        label= 'Track Temperature (°C)',
        s=100,
        zorder=2)

## Titlar etc..

ax.set_title('Average difference between Air and Track Temperature (°C) by Circuit', fontsize=16, weight='bold', pad=20)

ax.set_xlabel('Temperature (°C)', fontsize=12)
ax.set_ylabel('')


ax.grid(axis='x', color='gray', linestyle='--', alpha=0.5)

# sub title

ax.text(0.3, 1.01, 'Orange = Track Temperature, Blue = Air Temperature', transform=ax.transAxes, ha='center', fontsize=10, color='lightgray')
ax.text(0.8, 1.01, 'Data from 2023-2025 F1 sessions', transform=ax.transAxes, ha='center', fontsize=10, color='lightgray')
ax.text(0.5, -0.12, 'Longer line = Higher tyre degradation', transform=ax.transAxes,  ha='center')
```
