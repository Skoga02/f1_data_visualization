```python
import pandas as pd

df = pd.read_csv("drivers_2023_2025_final.csv")
df.head()
```

```python
df.columns
```

Top 3 teams vinster under år 2025

```python
df_2025 = df[df["year"] == 2025]
team_wins = df_2025.groupby("team")["wins"].sum()
team_wins = team_wins.sort_values(ascending=False)
team_wins_top3 = team_wins.head(3)
print(team_wins_top3)
```

Vilka länder har mest förare?

```python
df_unique = df.drop_duplicates(subset=["number"])
country_counts = df_unique["country"].value_counts()
print(country_counts.head(10))
```
