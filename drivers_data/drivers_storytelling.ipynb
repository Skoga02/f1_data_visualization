```python
import pandas as pd
import duckdb
import matplotlib.pyplot as plt
from constants import GRAY_1, GRAY_2, GRAY_3, BLUE_1

df = pd.read_csv("drivers_2023_2025_final.csv")
df.head()
```

```python
df_unique = df.drop_duplicates(subset=["number"])
country_counts = df_unique["country"].value_counts()
print(country_counts.head(10))
```

```python
plt.style.use("default")

ax = country_counts.plot(kind ="bar", ylabel ="", xlabel ="Countries", figsize=(6,5))

colors = [GRAY_1] * len(country_counts) 

max_index = country_counts.argmax()
colors[max_index] = "#E10600"  

ax.spines[["top","right"]].set_visible(False)
ax.spines [["left", "bottom"]].set_color(GRAY_2)

ax.tick_params(axis="both", colors =GRAY_2, length = 0)
ax.tick_params(axis="y", pad = 5, colors=GRAY_2)
ax.tick_params(axis= "x", pad = 3, colors=GRAY_2)

ax.set_title(
    "The UK dominates F1 with producing drivers",
    loc = "left",
    pad = 20,
    color = GRAY_2,
    fontweight="bold"
)

ax = country_counts.plot(
    kind="bar",
    ylabel="",
    xlabel="Countries",
    figsize=(10,6),
    color=colors
)
ax.set_xlabel("")

ax.set_yticks(range(0, int(country_counts.max()) + 1))

ax.set_xlabel("Countries", color = GRAY_2,)
ax.set_ylabel("Drivers", color=GRAY_2, rotation = 0)
ax.yaxis.set_label_coords(-0.05, 1.0)


for i, v in enumerate(country_counts.values):
    ax.text(i, v + 0.05, str(v), ha='center', color=GRAY_2, fontweight= "bold")

ax.set_xticklabels(country_counts.index, rotation=0)

#ax.yaxis.grid(True, linestyle="--", alpha=0.2)
ax.xaxis.grid(False)

fig = ax.get_figure()
fig.tight_layout()
fig.savefig("drivers_barchart.png")
```
